```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('desktop/data/auto-mpg.csv')
```

```python
df.head()
```

```python
#origin #1-usa, 2-Europe, 3-Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'
df['origin']=df['origin'].apply(origin)
```

```python
df.head(30)
```

```python
sns.set_style('whitegrid')
sns.pairplot(df[['mpg','weight','origin']],hue='origin',kind='reg',size=4)
```

```python
#piechart
cause=['tets','trftsf','vdcgc']
pc=[73,49,29]
explode=(0.15,0.15,0.15)
plt.figure(figsize=(5,5))
plt.pie(pc,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
#plt.axis('equal')
plt.title('state')
plt.legend()
plt.show()
```
