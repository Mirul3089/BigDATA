```python
final_df=train_df.append(test_df)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey']=pd.to_datetime(final_df['Date_of_Journey'])
```

```python
final_df['year']=pd.DatetimeIndex(final_df['Date_of_Journey']).year
final_df['month']=pd.DatetimeIndex(final_df['Date_of_Journey']).month
final_df['date']=pd.DatetimeIndex(final_df['Date_of_Journey']).day
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
#filling the null values by using columns mode
final_df['Route']=final_df['Route'].fillna(final_df['Route'].mode)
```

```python
#filling the null values by using columns mode
final_df['Total_Stops']=final_df['Total_Stops'].fillna(final_df['Total_Stops'].mode)
```

```python
final_df.info()
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x:x.split(":")[0])
final_df['Arrival_minutes']=final_df['Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_minutes']=final_df['Arrival_minutes'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Dept_hour']=final_df['Dep_Time'].apply(lambda x:x.split(":")[0])
final_df['Dept_minutes']=final_df['Dep_Time'].apply(lambda x:x.split(":")[1])
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_minutes']=final_df['Dept_minutes'].astype(int)
```

```python
final_df.head()
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4})
```

```python
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
#final_df['Duration_hour']=final_df['Duration'].str.split('h').str[0]
s=pd.to_timedelta(final_df['Duration'])
final_df['Duration_hours']=s.dt.components['hours']
final_df['Duration_minutes']=s.dt.components['minutes']
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_minutes1']=final_df['Duration_hours']*60
```

```python
final_df.info()
```

```python
final_df['Duration_minutes']=final_df['Duration_minutes1']+final_df['Duration_minutes']
```

```python
final_df.head()
```

```python
final_df.drop('Duration_minutes1',axis=1,inplace=True)
#axis=0 is to drop column
#axis=1 is to drop row
```

```python
final_df.drop()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```
